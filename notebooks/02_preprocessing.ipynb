### Handling `Lot Frontage`

- `Lot Frontage` was first imputed using the median value within each `Neighborhood`.
- During validation, three records (`GrnHill` and `Landmrk`) still had missing values because those neighborhoods contained no valid `Lot Frontage` observations.
- For these rare cases, the overall dataset median was used as a fallback.
- This hierarchical imputation preserves neighborhood-specific information whenever possible while ensuring no missing values remain.

```python
#check the data that has zero null values
# Fill missing Lot Frontage values using the median of each Neighborhood
#if there is no value then look into the median values

#  Calculate neighborhood medians
neighborhood_median = (
    df.groupby("Neighborhood")["Lot Frontage"]
      .median()
)

# Fill using neighborhood median
df["Lot Frontage"] = df["Lot Frontage"].fillna(
    df["Neighborhood"].map(neighborhood_median)
)

#Fill any remaining values using the overall median
overall_median = df["Lot Frontage"].median()

df["Lot Frontage"] = df["Lot Frontage"].fillna(overall_median)

#  Verify
print(df["Lot Frontage"].isnull().sum())
```
